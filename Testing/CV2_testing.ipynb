```python
# Importing Special Libraries

from deepface import DeepFace
import cv2
```

```python
img_1 = '/home/parker/Desktop/TInder_Project/Abbie_63952d667556000100d9c9be_3_1.jpeg'
img1 = cv2.imread(img_1)

plt.imshow(img1[:,:,::-1])
```

```python
face1 = DeepFace.detectFace(img1,
                            target_size=(224,224),
                            detector_backend= 'opencv'
)


plt.imshow(face1)
```

```python
obj1 = DeepFace.analyze(face1)
```
